```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from scipy.stats import ttest_rel
from seaborn import despine

from matplotlib import rcParams
rcParams['font.family'] = 'Helvetica'

from plotting_functions import *
```

```python
def cohen_d(means_pre, means_post):
    d_means = means_post-means_pre
    avg_d_means = np.mean(d_means)
    std_pre = np.std(means_pre)
    std_post = np.std(means_post)
    d = (avg_d_means)/(std_post-std_pre)
    return d
```

```python
# load features dataframe
data_path = '/Users/sydneysmith/Projects/ECT/NEW_ORGANIZATION/02-Data/AnalysisFeatures/all_features_bandpow4.csv'

exp_df = pd.read_csv(data_path)

# drop non-seizure WAR
exp_df = exp_df.drop(exp_df[exp_df['patients']=='WAR'].index)
```

```python
# get dataframe of means across all frontal electrodes
features = ['chans_exps', 'deltas', 'delta_pows', 'delta_band_pow'] #'alphas', 'alpha_pows', 'thetas', 'theta_pows', 'alpha_band_pow', 'theta_band_pow', 'l_pow', 'h_pow'
exp_df_mean = pd.DataFrame(exp_df.groupby(['patients', 'sessions', 'preposts'], as_index=False)[features].mean())

# filter out incomplete subj
count_df = pd.DataFrame(exp_df_mean.groupby(['patients'])['chans_exps'].count())
count_df_filt = count_df[count_df['chans_exps']>3]
complete_subj = list(count_df_filt.index)

filt_df_mean = exp_df_mean[exp_df_mean['patients'].isin(complete_subj)]

# get dataframe of matched pre/post deltas
delta_mean = filt_df_mean.dropna(how='any')
delt_df = pd.DataFrame(delta_mean.groupby(['patients', 'sessions'])['deltas'].count())
delt_df_filt = delt_df[delt_df['deltas']>1]
delt_subj, delt_ses = zip(*list(delt_df_filt.index))

delta_filt_mean = delta_mean[delta_mean['patients'].isin(delt_subj) & delta_mean['sessions'].isin(delt_ses)]
```

### Alpha CF change in one patient

```python
import mne
from neurodsp import spectral
from fooof import FOOOF
from fooof import analysis

from ECT_Functions import get_info, concatenate_annot
```

```python
pat = 'PAR'
ses = '1'

root = '/Users/sydneysmith/Projects/ECT/NEW_ORGANIZATION/02-Data/RawFIF/'
pre_path = root + '['+ pat + '] (Pre) Session ' + ses + ' (12.02.19)raw.fif'
post_path = root + '['+ pat + '] (Post) Session ' + ses + ' (12.02.19)raw.fif'

channels = [0,1,2,11,12,13]
labels = {0:'AF3', 1:'F7', 2:'F3', 11:'F4', 12:'F8', 13:'AF4'}

# FOOOF Settings
fm = FOOOF(peak_threshold=1.5, peak_width_limits=[1.00,12.0], aperiodic_mode='fixed')
```

```python
pre_psds = np.zeros(shape=(6,59))

fif = mne.io.read_raw_fif(pre_path, verbose=False)
annot = fif.annotations
fs = fif.info['sfreq']
bads = fif.info['bads']
#         ind_pre_chan = 0
#         ind_post_chan = 0
for ind, channel in enumerate(channels):
    if labels[channel] in bads:
        pass
    else: 
        data = concatenate_annot(fif, annot, channel, fs)
        output = spectral.compute_spectrum_welch(data, fs, avg_type='median', window='hann',
                                                 nperseg=fs*2, noverlap=fs, f_range=(1,30))
        spectra = output[1]
        freqs = output[0]

        fm.fit(freqs, spectra, freq_range=[1,30])
        peak_params = fm.peak_params_
        psd = fm.fooofed_spectrum_
        offset = fm.aperiodic_params_[0]
        # alpha peak
        alpha = analysis.periodic.get_band_peak(peak_params, band=(8.,14.))
        alpha_cf = alpha[0]
        # theta peak
        theta = analysis.periodic.get_band_peak(peak_params, band=(3.,8.))
        theta_cf = theta[0]
        exponent = fm.aperiodic_params_[1]
        adj_psd = psd - offset

#         patient, date, session, prepost, chan, hemi = get_info(file, channel)
        pre_psds[ind,:] = adj_psd
```

```python
post_psds = np.zeros(shape=(6,59))

fif = mne.io.read_raw_fif(post_path, verbose=False)
annot = fif.annotations
fs = fif.info['sfreq']
bads = fif.info['bads']
#         ind_pre_chan = 0
#         ind_post_chan = 0
for ind, channel in enumerate(channels):
    if labels[channel] in bads:
        pass
    else: 
        data = concatenate_annot(fif, annot, channel, fs)
        output = spectral.compute_spectrum_welch(data, fs, avg_type='median', window='hann',
                                                 nperseg=fs*2, noverlap=fs, f_range=(1,30))
        spectra = output[1]
        freqs = output[0]

        fm.fit(freqs, spectra, freq_range=[1,30])
        peak_params = fm.peak_params_
        psd = fm.fooofed_spectrum_
        offset = fm.aperiodic_params_[0]
        # primary oscillation
        cf = analysis.periodic.get_band_peak(peak_params, band=(3.,14.))
        cf = cf[0]
        # alpha peak
        alpha = analysis.periodic.get_band_peak(peak_params, band=(8.,14.))
        alpha_cf = alpha[0]
        # theta peak
        theta = analysis.periodic.get_band_peak(peak_params, band=(3.,8.))
        theta_cf = theta[0]
        exponent = fm.aperiodic_params_[1]
        adj_psd = psd - offset
        post_psds[ind,:] = adj_psd

#         patient, date, session, prepost, chan, hemi = get_info(file, channel)
```

```python
pre_psds = pre_psds[~np.all(pre_psds == 0, axis=1)]
post_psds = post_psds[~np.all(post_psds == 0, axis=1)]
```

```python
avg_pre= np.mean(pre_psds, axis=0)
avg_post = np.mean(post_psds, axis=0)
```

```python
plt.figure(figsize=(12,10))
sns.set_context('poster')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')

for s in post_psds:
    if s[1] == 0:
        pass
    else:
        plt.plot(freqs, s, color = 'orange', alpha=0.1, lw=4)
for s in pre_psds:
    if s[1] == 0:
        pass
    else:
        plt.plot(freqs, s, color = 'teal', alpha=0.1, lw=4)

plt.plot(freqs, avg_pre, color = 'teal', lw=10, label='Pre')
plt.plot(freqs, avg_post, color = 'orange', lw=10, label='Post')

plt.legend(prop={'size': 32})
        
plt.tight_layout()
despine()
```
