### Alpha CF change in one patient

```python
import mne
from neurodsp import spectral
from fooof import FOOOF
from fooof import analysis

from ECT_Functions import get_info, concatenate_annot
```

```python
pat = 'PAR'
ses = '1'

root = '/Users/sydneysmith/Projects/ECT/NEW_ORGANIZATION/02-Data/RawFIF/'
pre_path = root + '['+ pat + '] (Pre) Session ' + ses + ' (12.02.19)raw.fif'
post_path = root + '['+ pat + '] (Post) Session ' + ses + ' (12.02.19)raw.fif'

channels = [0,1,2,11,12,13]
labels = {0:'AF3', 1:'F7', 2:'F3', 11:'F4', 12:'F8', 13:'AF4'}

# FOOOF Settings
fm = FOOOF(peak_threshold=1.5, peak_width_limits=[1.00,12.0], aperiodic_mode='fixed')
```

```python
pre_psds = np.zeros(shape=(6,59))

fif = mne.io.read_raw_fif(pre_path, verbose=False)
annot = fif.annotations
fs = fif.info['sfreq']
bads = fif.info['bads']
#         ind_pre_chan = 0
#         ind_post_chan = 0
for ind, channel in enumerate(channels):
    if labels[channel] in bads:
        pass
    else: 
        data = concatenate_annot(fif, annot, channel, fs)
        output = spectral.compute_spectrum_welch(data, fs, avg_type='median', window='hann',
                                                 nperseg=fs*2, noverlap=fs, f_range=(1,30))
        spectra = output[1]
        freqs = output[0]

        fm.fit(freqs, spectra, freq_range=[1,30])
        peak_params = fm.peak_params_
        psd = fm.fooofed_spectrum_
        offset = fm.aperiodic_params_[0]
        # alpha peak
        alpha = analysis.periodic.get_band_peak(peak_params, band=(8.,14.))
        alpha_cf = alpha[0]
        # theta peak
        theta = analysis.periodic.get_band_peak(peak_params, band=(3.,8.))
        theta_cf = theta[0]
        exponent = fm.aperiodic_params_[1]
        adj_psd = psd - offset

#         patient, date, session, prepost, chan, hemi = get_info(file, channel)
        pre_psds[ind,:] = adj_psd
```

```python
post_psds = np.zeros(shape=(6,59))

fif = mne.io.read_raw_fif(post_path, verbose=False)
annot = fif.annotations
fs = fif.info['sfreq']
bads = fif.info['bads']
#         ind_pre_chan = 0
#         ind_post_chan = 0
for ind, channel in enumerate(channels):
    if labels[channel] in bads:
        pass
    else: 
        data = concatenate_annot(fif, annot, channel, fs)
        output = spectral.compute_spectrum_welch(data, fs, avg_type='median', window='hann',
                                                 nperseg=fs*2, noverlap=fs, f_range=(1,30))
        spectra = output[1]
        freqs = output[0]

        fm.fit(freqs, spectra, freq_range=[1,30])
        peak_params = fm.peak_params_
        psd = fm.fooofed_spectrum_
        offset = fm.aperiodic_params_[0]
        # primary oscillation
        cf = analysis.periodic.get_band_peak(peak_params, band=(3.,14.))
        cf = cf[0]
        # alpha peak
        alpha = analysis.periodic.get_band_peak(peak_params, band=(8.,14.))
        alpha_cf = alpha[0]
        # theta peak
        theta = analysis.periodic.get_band_peak(peak_params, band=(3.,8.))
        theta_cf = theta[0]
        exponent = fm.aperiodic_params_[1]
        adj_psd = psd - offset
        post_psds[ind,:] = adj_psd

#         patient, date, session, prepost, chan, hemi = get_info(file, channel)
```

```python
pre_psds = pre_psds[~np.all(pre_psds == 0, axis=1)]
post_psds = post_psds[~np.all(post_psds == 0, axis=1)]
```

```python
avg_pre= np.mean(pre_psds, axis=0)
avg_post = np.mean(post_psds, axis=0)
```

```python
plt.figure(figsize=(12,10))
sns.set_context('poster')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')

for s in post_psds:
    if s[1] == 0:
        pass
    else:
        plt.plot(freqs, s, color = 'orange', alpha=0.1, lw=4)
for s in pre_psds:
    if s[1] == 0:
        pass
    else:
        plt.plot(freqs, s, color = 'teal', alpha=0.1, lw=4)

plt.plot(freqs, avg_pre, color = 'teal', lw=10, label='Pre')
plt.plot(freqs, avg_post, color = 'orange', lw=10, label='Post')

plt.legend(prop={'size': 32})
        
plt.tight_layout()
despine()
```
