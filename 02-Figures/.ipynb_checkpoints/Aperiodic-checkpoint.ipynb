```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from scipy.stats import ttest_rel
from seaborn import despine

from matplotlib import rcParams
rcParams['font.family'] = 'Helvetica'

from plotting_functions import *
```

```python
# load features dataframe
data_path = '/Users/sydneysmith/Projects/ECT/NEW_ORGANIZATION/02-Data/AnalysisFeatures/all_features.csv'

exp_df = pd.read_csv(data_path)

# drop non-seizure WAR
exp_df = exp_df.drop(exp_df[exp_df['patients']=='WAR'].index)
```

```python
# get dataframe of means across all frontal electrodes
features = ['chans_exps', 'alphas', 'alpha_pows', 'thetas', 'theta_pows']
exp_df_mean = pd.DataFrame(exp_df.groupby(['patients', 'sessions', 'preposts'], as_index=False)[features].mean())

# filter out incomplete subj
count_df = pd.DataFrame(exp_df_mean.groupby(['patients'])['chans_exps'].count())
count_df_filt = count_df[count_df['chans_exps']>3]
complete_subj = list(count_df_filt.index)

filt_df_mean = exp_df_mean[exp_df_mean['patients'].isin(complete_subj)]
```

## 1 - Pre/Post changes 

### 1a -  Aperiodic (AP) exponent

#### AP exponent increases following a session

```python
plot_feature_prepost(filt_df_mean, feature='chans_exps')
despine()
plt.ylabel('aperiodic exponent')

plt.savefig('./Figures/AP_prepost.png')
```

AP exponent increase is significant

```python
means_pre = filt_df_mean[filt_df_mean['preposts']=='Pre']['chans_exps'].values
means_post = filt_df_mean[filt_df_mean['preposts']=='Post']['chans_exps'].values

print(ttest_rel(means_pre, means_post))
```

#### AP Exponent changes over sessions

```python
plot_feature_prepost_sessions(filt_df_mean, feature='chans_exps')
despine()
```

#### The change in aperiodic exponent does NOT significantly change over the course of treatment
 
no delta-of-deltas effect

```python
plot_feature_sessions(filt_df_mean, feature='chans_exps', prepost='diff')
despine()
```

#### Linear mixed effects model of AP Exponent change by session

```python
# Get dataframe with differences in features
df_diff = filt_df_mean.groupby(['patients','sessions'], as_index=False)[features].diff().dropna(0,how='all')
df_diff['patients'] = filt_df_mean['patients']
df_diff['sessions'] = filt_df_mean['sessions']
```

```python
md = smf.mixedlm("chans_exps ~ sessions", df_diff, groups=df_diff["patients"])
mdf = md.fit()
mdf.summary()
```

## 4 - Hemispheric differences in alpha oscillations

```python
# hemispheric dataframe
hemi_df_mean = pd.DataFrame(exp_df.groupby(['patients', 'sessions', 'preposts', 'hemis'], as_index=False)[features].mean())

# filter out incomplete subj
# count_df = pd.DataFrame(hemi_df_mean.groupby(['patients','sessions', 'preposts'])['alphas'].count())
# count_df_filt = count_df[count_df['alphas']%2 == 0]
# complete_subj = list(count_df_filt.index)

filt_hemi_df_mean = hemi_df_mean[hemi_df_mean['patients'].isin(complete_subj)]
```

```python
filt_hemi_df_mean
```

```python
# Dataframe with difference between hemispheres for each feature
df_diff_hemi = filt_hemi_df_mean.groupby(['patients', 'sessions', 'preposts'], as_index=False)[features].diff().dropna(0,how='all')
df_diff_hemi['patients'] = filt_hemi_df_mean['patients']
df_diff_hemi['sessions'] = filt_hemi_df_mean['sessions']
df_diff_hemi['preposts'] = filt_hemi_df_mean['preposts']
```

```python
plot_hemi_feature_prepost(filt_hemi_df_mean, feature='chans_exps')
despine()
```

This doesnt look like much

```python
# Does the differnce between the exponents of each hemisphere change after ECT? 
plot_feature_prepost(df_diff_hemi, feature='chans_exps')
despine()
```

```python
#apow_df_diff_hemi = df_diff_hemi[df_diff_hemi['alpha_pows'].notnull()]
count_df = pd.DataFrame(df_diff_hemi.groupby(['patients'])['chans_exps'].count())
count_df_filt = count_df[count_df['chans_exps']%2 == 0]
complete_subj = list(count_df_filt.index)
ap_df_diff_hemi = df_diff_hemi[df_diff_hemi['patients'].isin(complete_subj)]

means_pre = ap_df_diff_hemi[ap_df_diff_hemi['preposts']=='Pre']['chans_exps'].values
means_post = ap_df_diff_hemi[ap_df_diff_hemi['preposts']=='Post']['chans_exps'].values

print(ttest_rel(means_pre, means_post))
```

```python
## stats for hemi stuff

means_pre_left = filt_hemi_df_mean[(filt_hemi_df_mean['preposts']=='Pre') &
                                   (filt_hemi_df_mean['hemis']=='Left')]['chans_exps'].values
means_pre_right = filt_hemi_df_mean[(filt_hemi_df_mean['preposts']=='Pre') &
                                    (filt_hemi_df_mean['hemis']=='Right')]['chans_exps'].values
means_post_left = filt_hemi_df_mean[(filt_hemi_df_mean['preposts']=='Post') &
                                    (filt_hemi_df_mean['hemis']=='Left')]['chans_exps'].values
means_post_right = filt_hemi_df_mean[(filt_hemi_df_mean['preposts']=='Post') &
                                     (filt_hemi_df_mean['hemis']=='Right')]['chans_exps'].values

print('Pre-Right vs. Pre-Left: ' + str(ttest_rel(means_pre_right, means_pre_left)))
print('Post-Right vs. Post-Left: ' + str(ttest_rel(means_post_right, means_post_left)))
print('Pre-Left vs. Post-Left: ' + str(ttest_rel(means_pre_left, means_post_left)))
print('Pre-Right vs. Post-Right: ' + str(ttest_rel(means_pre_right, means_post_right)))
```

```python
plot_hemi_feature_prepost_sessions(filt_hemi_df_mean, feature='chans_exps')
despine()
```
